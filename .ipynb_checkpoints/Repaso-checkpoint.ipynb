# Datatype

1. Crea una variable entero con el valor 25, un flotante con 10.5 y un booleano con True. Imprime el tipo de cada uno usando type().

```python
x=25
print(x)
print(type(x))

y=10.5
print(y)
print(type(y))

z=True
print(z)
print(type(z))
```

2. Define una lista que contenga tres números, dos cadenas de texto y un booleano. Accede al segundo elemento.

```python
lista=[5, 3, 1, 'Pera', 'Piña', True]
print(lista)
print(type(lista))
print(lista[1])
print(type(lista[1]))
```

3. Convierte el número entero 100 a una cadena de texto y luego a un número de punto flotante.

```python
x=100
print(x)
print(type(x))

print(str(x))
print(type(str(x)))

print(float(x))
print(type(float(x)))

x= str(x)
print(x)

x= float(x)
print(x)
```

4. Crea una cadena de texto con tu nombre y utiliza el método para convertirla toda a mayúsculas.

```python
x='Judith'
print(x)
print(x.upper())
```

5. Dada la lista numeros = [10, 20, 30, 40], utiliza slicing para obtener solo los elementos del medio (20 y 30).

```python
numeros=[10, 20, 30, 40]
num_medio=numeros[1:3:1]
print(num_medio)
```
