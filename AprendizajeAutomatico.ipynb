# Metodología
## Introducción

**aprendizaje automático**

-  Desarrollar algoritmos que permiten a las computadoras reconocer patrones, hacer predicciones y tomar decisiones basadas en datos, mejorando su rendimiento con la experiencia acumulada. A diferencia de los programas tradicionales con reglas fijas, los sistemas de aprendizaje automático ajustan sus parámetros internos para minimizar errores y optimizar resultados.

**machine learning**

- "... rama de conocimiento en la que se abordan algoritmos que son capaces de crear modelos de
conocimiento abstractos a partir de históricos de datos … naturaleza empírica"

**data mining**

- "Conjunto de procesos, métodos y técnicas que conducen a la extracción de conocimiento a partir de bases de datos".

## Metodología CRISP-DM y CRISP-ML(Q)

Para llevar a cabo proyectos de forma racional y estructurada, se presentan marcos de referencia como **CRISP-ML(Q)**, que consta de las siguientes fases:

1. Comprensión del negocio y de los datos.
2. Ingeniería de datos (Preparación de datos).
3. Ingeniería de modelos de Machine Learning.
4. Garantía de calidad para aplicaciones de ML.
5. Despliegue.
6. Monitorización y mantenimiento.

# Detalles de las Fases de la Metodología

## Comprensión del Negocio
- Determinar objetivos empresariales.
- Definir objetivos de minería de datos.
- Evaluar la situación actual.
- Elaborar un plan de proyecto.

### Comprensión de los Datos
- Recolectar datos iniciales.
- Describir los datos:
  - Número de filas.
  - Número de columnas.
  - Significado de los atributos.
- Explorar los datos mediante estadística descriptiva.
- Verificar la calidad:
  - Valores nulos.
  - Valores fuera de rango.
  - Inconsistencias.

## Preparación de los Datos
- Seleccionar datos relevantes.
- Limpiar datos.
- Construir nuevos atributos.
- Integrar conjuntos de datos.
- Formatear datos para el modelado.

## Modelado
- Seleccionar algoritmos.
- Generar planes de prueba.
- Construir modelos.
- Evaluar resultados técnicos iniciales.

## Evaluación de Negocio
- Verificar si los modelos cumplen los criterios de éxito empresarial.
- Resumir hallazgos.
- Decidir si:
  - Se itera el proceso.
  - Se pasa a explotación.

## 1.9. Despliegue
- Transformar el conocimiento en acciones.
- Planificar la implementación.
- Monitorizar el rendimiento.
- Elaborar el informe final del proyecto.


# Tipos de Aprendizaje

## Aprendizaje Supervisado

Se utiliza cuando existe una variable objetivo conocida. Se utilizan etiquetas.

### Regresión
- La variable de salida es numérica.
- Produce valores continuos.

**Ejemplos:**
- Predecir ventas.
- Estimar precios.
- Calcular temperatura futura.

### Clasificación
- La variable de salida es categórica.
- Produce etiquetas o categorías.

**Ejemplos:**
- Aprobar / Reprobar.
- Enfermo / Sano.
- Spam / No Spam.

## Aprendizaje No Supervisado

Se utiliza cuando no existe una variable objetivo. No se utilizan etiquetas.

### Clustering
- Agrupa datos similares.
- Permite dividir problemas complejos.
- Facilita el análisis independiente de cada grupo.

**Aplicaciones:**
- Segmentación de clientes.
- Agrupación de productos.
- Detección de patrones ocultos.

### Reglas de Asociación
- Descubren relaciones entre variables.
- Utilizan relaciones antecedente → consecuente.

**Aplicaciones:**
- Sistemas de recomendación.
- Análisis de compras.
- Estudio de dependencias entre datos.

## Aprendizaje por Refuerzo

No utiliza datos etiquetados.
Basado en la interacción entre:
- Agente.
- Entorno.

El aprendizaje ocurre mediante:
- Recompensas.
- Penalizaciones.
- Retroalimentación (feedback).

**Aplicaciones:**
- Robótica.
- Videojuegos.
- Vehículos autónomos.


# Estructura de Datos

La estructura fundamental es la **tabla de datos** o **dataset**.
- “La tabla de datos es una estructura en dos dimensiones (con filas y columnas), al igual que las matrices en
matemáticas. No obstante, a diferencia de las matrices, las tablas de datos pueden contener diferentes tipos de
datos (números, cadenas, fechas…). A pesar de ello, cada columna de una tabla de datos debe contener solo
datos del mismo tipo. ”

# Limpieza de Datos

La limpieza de datos consiste en eliminar defectos y adaptar los datos para que los patrones sean más claros durante el modelado.

Puede aplicarse a nivel de:

- Fila.
- Columna.
- Matriz completa.

## Normalización y Estandarización

### Normalización

Transforma atributos numéricos a un rango continuo mediante homotecia.

**Objetivo:**
- Escalar datos.
- Evitar que atributos con valores grandes dominen el modelo.

### Estandarización

Ajusta los valores para que tengan:

- Media = 0
- Desviación estándar = 1

También conocida como: Z-Score

## Detección de Outliers

Identificación de observaciones que presentan comportamientos significativamente diferentes respecto al resto de los datos.

### Elementos Detectados

| Tipo |
|--------|
| Envolvente elíptica: únicamente con datos de distribución normal. Trabaja a nivel de matriz.|
| Isolation Forest: Trabaja a nivel de matriz. |
| Box-Whiskers: Trabaja a nivel de atributo. |

### Beneficios

- Mejora la calidad de los modelos.
- Reduce errores de predicción.
- Disminuye el impacto de datos erróneos.
