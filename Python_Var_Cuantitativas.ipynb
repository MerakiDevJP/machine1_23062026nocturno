```python
# Creación aleatoria de datos
import numpy as np
X = np.array([0.5, 23, 0.3, 4.5, 0.3, 0.5,
-28, -50, 60, -100, -10, -11, 13, 19, 1, 9]) #arreglo unidimensional - datos de entrada
print(X)
print("==============================================================================")
display(X)
```

```python
# Medidas de tendencia central
import numpy as np
from scipy import stats
'''
Mientras que NumPy proporciona la base para el manejo de arreglos numéricos (arrays), 
SciPy construye sobre ella para ofrecer una colección masiva de algoritmos y herramientas matemáticas de alto nivel.
'''
media = np.mean(X)
mediana = np.median(X)
moda , _ = stats.mode(X) 
'''el guion bajo (_) es una convención de nomenclatura 
utilizada para asignar un valor que no planeas utilizar posteriormente en tu código.
'''
print('media: ', media)
print('mediana: ', mediana)
print('moda :', moda)

print("==============================================================================")

display('media: ', media)
display('mediana: ', mediana)
display('moda :', moda)
```

```python
# Medidas de posición
import numpy as np
import matplotlib.pyplot as plt

Q1 = np.percentile(X, 25)
Q3 = np.percentile(X, 75)
RIC = Q3 - Q1 #es la línea amarilla o naranja Indicador intercuartil
lim_inf = Q1 - 1.5*RIC
lim_sup = Q3 + 1.5*RIC

print('RIC :', RIC)
print('limite inferior :', lim_inf)
print('limite superior :', lim_sup)

print("==============================================================================")

display('RIC :', RIC)
display('limite inferior :', lim_inf)
display('limite superior :', lim_sup)

#hay mucha distancia entre los datos debe ser analizada para ser normalizada
#poniendo todos los datos dentro de una caja entre 0 y 1
```

```python
# Dibuja el diagrama de box and whiskers y determina visualmente el número de outliers
plt.boxplot(X) #caja de bigotes
plt.grid()
plt.title("Diagrama de caja y bigotes")
plt.show()
'''
# outlayer - a ser eliminado
# límite superior
# Q3
# RIC
# Q1
# límite inferior
# outlayer - a ser eliminado
# outlayer - a ser eliminado
'''
```

```python

import matplotlib.pyplot as plt

# Definir propiedades de estilo
boxprops = dict(facecolor='skyblue', color='darkblue')      # Color de la caja
medianprops = dict(color='red', linewidth=2)               # Color de la línea de mediana
whiskerprops = dict(color='gray', linestyle='--')          # Color de los bigotes
flierprops = dict(marker='o', markerfacecolor='orange')    # Color de los outliers

plt.figure(figsize=(8, 6))

# Usamos patch_artist=True para poder rellenar el color
plt.boxplot(X, patch_artist=True, 
            boxprops=boxprops, 
            medianprops=medianprops,
            whiskerprops=whiskerprops,
            flierprops=flierprops)

plt.grid(axis='y', linestyle=':', alpha=0.7)
plt.title("Diagrama de caja y bigotes (Personalizado)")
plt.show()
```

```python
# Medidas de dispersión
desvest = np.std(X)
print('Desviación típica: ', desvest)
print('Varianza: ', desvest**2)

print("==============================================================================")

display('Desviación típica: ', desvest)
display('Varianza: ', desvest**2)
```

```python
# Medidas de distribución
asimetria = stats.skew(X)
curtosis = stats.kurtosis(X, fisher=True) # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kurtosis.html
print('Asimetría ', np.round(asimetria, 4))
print('Curtosis ', np.round(curtosis, 4))
if curtosis>0:
    print('Distribución Leptocúrtica')
elif curtosis<0:
    print('Distribución Platicúrtica')
else:
    print('Distribución Mesocúrtica')

print("=====================================================")

display('Asimetría ', np.round(asimetria, 4))
display('Curtosis ', np.round(curtosis, 4))
if curtosis>0:
    display('Distribución Leptocúrtica')
elif curtosis<0:
    display('Distribución Platicúrtica')
else:
    display('Distribución Mesocúrtica')
```

```python
# EJERCICIO EXTRA SOBRE EL DATASET DE IRIS
# Extraer la media y la desviación típica de cada variable en función de la clase y representarlo en un gráfico (plt.errorbar)
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets

# 1. Se carga el dataset y se prepara como DataFrame
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Se añade la columna de la especie (clase)
df['species'] = iris.target_names[iris.target]

# 2. Se calcula la media y desviación estándar agrupando por especie
stats_iris = df.groupby('species').agg(['mean', 'std'])

# 3. Representación gráfica con las variables
# Se configurar (2x2)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten() # Aplanar para iterar fácilmente

# 3. Se dibuja cada variable
for i, col in enumerate(iris.feature_names):
    means = stats_iris[col]['mean']
    stds = stats_iris[col]['std']
    
    axes[i].errorbar(x=stats_iris.index, y=means, yerr=stds, 
                     fmt='o', capsize=5, color='teal', ecolor='salmon', elinewidth=2)
    axes[i].set_title(col)
    axes[i].set_ylabel('cm')
    axes[i].grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()

resultados = {
    "Sepal Length": {
        "obs": "Las medias están algo separadas, pero las barras de error se solapan un poco entre Versicolor y Virginica.",
        "int": "Es un indicador útil, pero no definitivo por sí solo para clasificar."
    },
    "Sepal Width": {
        "obs": "Las medias están muy juntas y las barras de error se superponen significativamente.",
        "int": "Tiene poco poder discriminatorio; no es un buen rasgo para distinguir especies."
    },
    "Petal Length": {
        "obs": "Es el gráfico más claro; medias muy distantes y barras que apenas se tocan.",
        "int": "Es la variable más importante y con altísimo poder predictivo."
    },
    "Petal Width": {
        "obs": "Similar a la longitud, muestra buena separación entre grupos.",
        "int": "Variable altamente relevante; suficiente para clasificar casi perfectamente."
    }
}

print("Resultados del Análisis:\n" + "="*30)
for i, (nombre, info) in enumerate(resultados.items(), 1):
    print(f"{i}. {nombre}")
    print(f"Observación: {info['obs']}")
    print(f"Interpretación: {info['int']}\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets

# 1. Se preparan los datos
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]
stats = df.groupby('species').agg(['mean', 'std'])

# 2. Se genera el gráfico
plt.figure(figsize=(10, 6))
colores = {'setosa': 'tab:blue', 'versicolor': 'tab:orange', 'virginica': 'tab:green'}

for especie in stats.index:
    medias = stats.loc[especie, (slice(None), 'mean')]
    desviaciones = stats.loc[especie, (slice(None), 'std')]
    plt.errorbar(x=iris.feature_names, y=medias, yerr=desviaciones, 
                 fmt='-o', label=especie, capsize=5, color=colores[especie])

plt.title("Media y desviación estándar de cada variable por especie (Iris)")
plt.xticks(rotation=15)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# 3. Se realiza el análisis automático de resultados
print("\n" + "="*50)
print("ANÁLISIS DE LOS RESULTADOS (Basado en la separación de medias)")
print("="*50)

for col in iris.feature_names:
    medias_col = stats[col]['mean']
    rango = medias_col.max() - medias_col.min()
    
    print(f"\nVariable: {col}")
    if rango > 1.5:
        print("-> Interpretación: ALTO poder discriminatorio. Las especies presentan medias muy distantes, facilitando la clasificación.")
    else:
        print("-> Interpretación: BAJO poder discriminatorio. Las especies se solapan significativamente; esta variable aporta poca información única.")
    print(f"   (Diferencia máxima observada: {rango:.2f} cm)")
```
