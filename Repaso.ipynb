1. Escribe un programa que pida un número y determine si es positivo, negativo o cero usando if, elif y else.

```python
print("Para saber el tipo de número")
x=int(input("Ingrese un número: "))
if x > 0:
    print("Es positivo")
elif x < 0:
    print("Es negativo")
else:
    print("Es cero")
```


6. Escribe un programa que utilice un bucle for para calcular la suma de todos los números en una lista dada.

7. Crea un bucle while que pida al usuario introducir texto hasta que escriba la palabra "salir".

```python
print("Para suspender el sistema debes escribir esta palabra: salir")
texto = input("Ingresa el texto")

while texto != 'salir':
    texto = input("Ingresa el texto")
    
print("Sistema suspendido")
```

8. Utiliza la sentencia break dentro de un bucle para detener la ejecución cuando el contador alcance el número 5.

```python
contador = 1

while contador <= 10:
    print(contador)

    if contador == 5:
        print("Límite alcanzado")
        break
        
    contador += 1
```

9. Escribe un programa que determine si un año es bisiesto (divisible por 4, pero no por 100, o divisible por 400).

```python
anio = int(input("Ingrese un año: "))

if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
    print(f"El año {anio} es bisiesto.")
else:
    print(f"El año {anio} no es bisiesto.")
```

10. Utiliza continue en un bucle for para saltarte la impresión del número 3 en una secuencia del 1 al 5.

```python
for i in range(1, 6):

    if i == 3:
        continue  # Se salta el resto del código en esta iteración
    print(i)
```
