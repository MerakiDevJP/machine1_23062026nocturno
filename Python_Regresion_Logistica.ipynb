```python
from sklearn import metrics

def evaluacion(y_true, y_pred, metricas):
    resultados = {}
    for nombre, scorer in metricas.items():
        if isinstance(scorer, str):
            # Manejo de strings (métricas estándar de sklearn)
            if scorer == 'neg_mean_absolute_error':
                resultados[nombre] = metrics.mean_absolute_error(y_true, y_pred)
            elif scorer == 'r2':
                resultados[nombre] = metrics.r2_score(y_true, y_pred)
        else:
            # Manejo de objetos make_scorer
            # Usamos el objeto directamente como una función si es un 'callable'
            try:
                # Intenta ejecutar el scorer directamente
                resultados[nombre] = scorer(None, y_true, y_pred)
            except:
                # Si falla, es probable que sea una función simple definida en tu diccionario
                resultados[nombre] = scorer(y_true, y_pred)
    return resultados
```
