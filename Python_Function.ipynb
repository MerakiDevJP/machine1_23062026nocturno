# Funciones
## ¿Qué son?
- Son un mecanismo que te ofrece el lenguaje para agrupar conjuntos de instrucciones de manera que se puedan ejecutar cuando el programador considere oportuno.
- Las funciones van identificadas por un nombre.
- Se debe usar este nombre para invocar (ejecutar) el código de la función.
- Las funciones calculan un resultado (output) en base a unos parámetros de entrada (input).
- En cada invocación de la función, los parámetros de entrada pueden variar.

### Ejemplo

```python
def sumar(x, y):
    suma=x+y
    return suma
print(sumar(4,5))
print(sumar(-2,0))
```

## Paso de parámetros
- Tipos simples (inmutables) por valor: int, float, string.<br>
    – El efecto es como si se creara una copia dentro de la función, aunque realmente no es esto lo que ocurre.<br>
    – Los cambios dentro de la función no afectan fuera.

```python
def cambiar_valor(x):
    x += 3
    print('Dentro de la función:', x) # x ahora referencia a un nuevo objeto (el 3), pero esto no afecta a la variable de fuera de la función, que sigue apuntando al 2.
    return
a = 2

cambiar_valor(a)
print('Fuera de la función:', a)
```

- Tipos complejos (mutables) por referencia: list, set, dictionary.<br>
    – Dentro de la función se maneja el mismo objeto que se ha pasado desde fuera.
    – Los cambios dentro de la función sí afectan fuera.

```python
def anyadir_2_a(x):
    x.append(4)
    
lista = [0, 1,3]
anyadir_2_a(lista)
print(lista)
```

```python
# with tuples
def anyadir_2_a(x):
    print(id(x))
    return
lista = (0, 1)
print(id(lista))

anyadir_2_a(lista)
print(lista)
print(type(lista))
```

- Si quiero modificar un objeto que es de un tipo básico, los cuales se pasan por valor, puedo simplemente devolver el resultado de la función y hacer una asignación:

```python
def multiplicar_por_2(x):
    x = x * 2
    return x
    
a = 3
print(a)
a = multiplicar_por_2(a)
print(a)
```

- Si quiero que no se modifique un objeto de un tipo complejo, los cuales se pasan por referencia,
puedo pasar una copia a la función:

```python
def anyadir_2_a(x):
    x.append(2)
    print('Dentro de la función: ', x)
    
lista = [0, 1]
anyadir_2_a(lista.copy())
print('Fuera de la función: ', lista)
```

```python
# Alternativa para obtener una copia: slicing.
lista = [0, 1]
anyadir_2_a(lista[:])
print('Fuera de la función: ', lista)
```

- Si la lista tiene sublistas anidadas hay que hacer un deep copy

```python
import copy

lista = [2, 4, 16, 32, [34, 10, [5,5]]]
copia = lista.copy()
copia = copy.deepcopy(lista)

copia[0] = 454
copia[4][2][0] = 64

print(lista)
print(copia)

print(f"{id(lista)} - {id(copia)}")
print(f"{id(lista[4])} - {id(copia[4])}")
```

## Recursividad
- Funciones que se llaman a sí mismas

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(10))

def fact(n):
    a = 1
    for i in range(1, n+1):
        a *= i
    return a

print(fact(10))
```

```python
fact = lambda x: 1 if x == 0 else x * fact(x-1)

fact(10)
```

## Funciones Lambda
- Es posible definir funciones anónimas (sin nombre).
- Lambdas son expresiones; por lo tanto, puede aparecer en lugares donde una sentencia def
no puede (por ejemplo, dentro de una lista o como parámetro de una función).
- Útiles cuando se quiere pasar una función como argumento a otra.

Formato general:<br>
`lambda <lista de argumentos> : <valor a retornar>`<br>
- Importante: `<valor a retornar>` no es un conjunto de instrucciones. Es simplemente una
expresión *return* que omite esta palabra.
- Las funciones definidas con *def* son más generales:<br>
– Cualquier cosa que implementes en un lambda, lo puedes implementar como una función convencional con *def*, pero no viceversa.

```python
#Ejemplo: función suma.

def suma(x, y, z):
    return x + y + z

print(suma(1, 2, 3))

# Ejemplo: función suma como expresión lambda.

suma2 = lambda x, y, z : x + y + z

print(suma2(1, 2, 3))
```

```python
# La funcion 'sort' puede recibir una función optional como parámetro.
# Esta función 'key' se invoca para cada elemento de la lista antes de realizar las comparaciones.
# Ejemplo: ordenar strings por número de carácteres.

cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

cities.sort()
print(cities)

cities.sort(key = lambda x : len(x))
print(cities)

cities.sort(key = len)
print(cities)

def longitud(x):
    return len(x)

cities.sort(key = longitud)
print(cities)
```

```python
# Mismo ejemplo sin lambda

cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

def num_caracteres(x):
    return len(x)

cities.sort(key = num_caracteres)
print(cities)
```

- Nos podemos guardar una referencia para utilización repetida.

```python
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

f = lambda x : len(x)

for s in cities:
    print(f(s))
```

-  Las expresiones lambda pueden formar parte de una lista.

```python
# Lista de lambdas para mostrar las potencias de 2.

pows = [lambda x: x ** 0,
        lambda x: x ** 1,
        lambda x: x ** 2,
        lambda x: x ** 3,
        lambda x: x ** 4]

for f in pows:
    print(f(2))
```

- Diccionarios se pueden utilizar como ‘switch’ en otros lenguajes
- Ideal para escoger entre varias opciones


```python
def switch_dict(operator, x, y):
    # crea un diccionario de opciones y selecciona 'operator'
    return {
        'add': lambda: x + y,
        'sub': lambda: x - y,
        'mul': lambda: x * y,
        'div': lambda: x / y,
    }.get(operator, lambda: None)() # retorna None si no encuentra 'operator'

print(switch_dict('add',2,2))
print(switch_dict('div',10,2))
print(switch_dict('mod',17,3))
```

## Scopes
- La localización de una asignación a una variable en el código determina desde donde puedes acceder a esa variable.
- Por ejemplo, una variable asignada dentro de una función sólo es visible dentro de esa función.
- El alcance de la visibilidad de una variable determina su *scope*.
- El término *scope* hace referencia a un espacio de nombres (*namespace*). Por ejemplo, una función establece su propio namespace.

```python
X = 10

def func():
    X = 20 # Local a la función. Es una variable diferente, no visible fuera de 'func'

    def func_int():
        X = 30
        print(X)
    
    func_int()

func()
```

**Regla LEGB**

Dada una función F, tenemos los siguientes *scopes*:
- **L**ocal: variables locales a F.
- **E**nclosing: variables localizadas en funciones que contienen a (o por encima de) F.
- **G**lobal: variables definidas en el módulo (fichero) que no están contenidas en ninguna función. Este scope no abarca más de un fichero.
- **B**uilt-ins: proporcionadas por el lenguaje.<br>

La resolución de nombres ocurre de abajo a arriba.

```python
# Global (X y func)
X = 99

def func(Y): # Local (Y y Z)
    Z = X + Y
    return Z

print(func(1))
```

```python
X = 1

def func():
    X = 2 # X es una variable diferente

func()
print(X)
```

- La sentencia *global* nos permite modificar una variable global desde dentro de una función.

```python
X = 1

def func():
    global X
    X = 2

func()
print(X)
```

- No hay necesidad de usar *global* para referenciar variables. Sólo es necesario para modificarlas.

```python
y, z = 1, 2

def todas_globales():
    global x
    x = y + z # Las tres variables son globales.

todas_globales()
print(x)
print(y)
print(z)
```

```python
X = 77

def f1():
    X = 88 # Enclosing scope (para f2)
    def f2():
        print(X)
    f2()

f1()
```

**Closures**
- Las funciones pueden recordar su enclosing scope, independientemente de si éstos continuan existiendo o no.


```python
def f1():
    X = 88 # Enclosing scope (para f2)
    def f2():
        print(X)
    return f2 # Devuelve f2, sin invocarla

accion = f1()
accion()
```

- Sentencia *nonlocal* para modificar variables que no son locales, pero tampoco globales.

```python
def f1():
    contador = 10
    def f2():
        nonlocal contador
        contador += 1
        print(contador)
    return f2

accion = f1()
accion()
accion()
accion()
```


2. Escribe una función que reciba números como entrada y devuelva la suma de los mismos. La función debe ser capaz de recibir una cantidad indeterminada de números. La función no debe recibir directamente ningún objeto complejo (lista, conjunto, etc.).

```python
# 1. Definición de la función (recibe parámetros individuales)
def sumar_numeros(*args): # *args recibe una cantidad indeterminada de argumentos y los convierte en tupla
    return sum(args) # Retorna la suma total de todos los elementos recibidos

# 2. Definición de las variables para controlar el bucle
numeros_ingresados = [] # Inicializa una lista vacía para acumular los números
continuar = True        # Define una variable booleana para mantener el bucle activo

print("Ingresa un número. Si deseas terminar escribe 'fin'.")
print("----------------------------------------------------") 

# 3. Desarrollo del bucle para pedir números de forma indeterminada
while continuar: # Inicia mientras 'continuar' sea True
    entrada = input("Ingrese un número: ") #
    print("----------------------------------------------------") 
    
    if entrada.lower() == 'fin': # Esta función devuelve los caracteres en minúsculas
        continuar = False # Cambia el estado para terminar el bucle
    else: # Si no es 'fin', se procesa el número
        try: # Inicio control de errores por si no es un número válido
            num = int(entrada) # Convierte el string a int
            numeros_ingresados.append(num) # Agrega el número convertido a la lista de acumulados
            
            # Muestra el número recién ingresado
            print(f"Número ingresado: {num}") # Concatena la respuesta
            
            # 3. Llamada a la función pasándole todos los números acumulados
            # El * pasa los elementos de la lista como argumentos individuales a la función
            suma_total = sumar_numeros(*numeros_ingresados) # Calcula la suma llamando a la función
            print(f"Suma acumulada hasta ahora: {suma_total}") # Concatena la respuesta con la suma acumulada
            print("----------------------------------------------------") 
        except ValueError: # Captura el error si el usuario ingresa algo que no es un número entero
            print("Por favor, ingresa un número válido o 'fin'.") # Notifica al usuario sobre el error
print("----------------------------------------------------") 
print(f"La suma final es: {sumar_numeros(*numeros_ingresados)}") # Concatena el resultado total de la suma
print("----------------------------------------------------") 
```


9. Escribe una función que compruebe si un string dado es un palíndromo. Un palíndromo es una secuencia de caracteres que se lee igual de izquierda a derecha que de derecha a izquierda.
Por ejemplo, la función devolverá True si recibe el string “reconocer” y False si recibe el string “python”.


```python
# 1. Definición de la función
def es_palindromo(cadena):
    # Se normaliza eliminando espacios y convirtiendo a minúsculas
    cadena_limpia = cadena.replace(" ", "").lower()
    # Se compara la cadena limpia con su versión invertida
    return cadena_limpia == cadena_limpia[::-1]

# 2. Se prueba la función y se muestra el resultado
print(es_palindromo("reconocer")) # Devuelve: True
print(es_palindromo("python"))    # Devuelve: False
print(es_palindromo("Anita lava la tina")) # Devuelve: True debido a la limpieza de espacios
```
