# .plot()
Múltiples parámetros
`plt.plot(x, y, style, label, linewidth=1, alpha=1)`

- ‘style’: cadena de texto con formato (color, tipo…) [color][marker][line]
- ‘label’: referencia para leyenda
- ‘linewidth’: brocha
- ‘alpha’: transparencia
Otros: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
y_data = np.random.randint(100,size=(10))
x_data = np.arange(10)

fig,axis = plt.subplots(1,1)
axis.plot(x_data,y_data,'gh--',label='dollars',linewidth=1,alpha=0.6) #creación del gráfico
axis.legend(loc='best') # representar la leyenda
plt.show()
```

```python
# Horizontal bar
fig, ax = plt.subplots()

# Example data
people = ('PhD', 'MSc', 'BSc', 'N/A')
y_pos = np.arange(len(people))
percentage = [5,45,25,25]

# print(y_pos)
ax.barh(y_pos, percentage, align='center',
        color='green', edgecolor='red')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)

#ax.invert_yaxis() # labels read top-to-bottom
ax.set_xlabel('Percentage of workforce')
ax.set_title('Level of education in the company')
plt.show()
```

```python
ls_count = (284487, 244560, 208493, 196764)
cities = ('Madrid', 'Barcelona', 'Sevilla', 'Valencia')
data = np.array(ls_count)/1000
y_pos = np.arange(data.size)

width = 0.35 # the width of the bars
opacity = 0.7

fig, ax = plt.subplots()
rects1 = ax.barh(y_pos, data, width, alpha=opacity)

rects1[3].set_color('#872f29')
rects1[2].set_color('#ffcb59')
rects1[1].set_color('#d6ee39')
rects1[0].set_color('#6ade30')

ax.set_yticks(y_pos)
ax.set_yticklabels(cities)

# add some text for labels, title and axes ticks
ax.set_ylabel('Ciudades')
ax.set_xlabel('Km de carril bici')

ax.legend((rects1[0], rects1[1], rects1[2], rects1[3]), cities)

plt.show()
```

# frame.plot()
`frame.plot(x,y,label=<Series>,ax,style,kind,xticks,yticks,title,subplots)`
- (solo para Series) label: referencia para la leyenda
- ax: subfigura en la que dibujar los datos
- style: estilo de la línea
- kind: tipo de gráfico (’bar’, ‘pie’, ‘hist’, ‘area’ ‘line’, ‘barh’, ‘density’, ‘kde’).
- xticks y yticks: valores en los ejes X e Y
- title: string como título
- (solo para DataFrame) subplots: bool si se desean subfiguras separadas para cada columna.
- (solo para DataFrame) ‘x’ e ‘y’ se pueden utilizar para seleccionar columnas para cada eje

[Lista completa](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html) de argumentos

```python
#scatter de clases, coloreando segun clase en matplotlib
iris = sns.load_dataset('iris')

# preparacion de los datos
data = []
colors = ["red", "green", "blue"]
groups = ["setosa", "versicolor", "virginica"]
columns = ["petal_length", "petal_width"]
for i in groups:
    # busqueda por clase definida
    data.append(iris[iris["species"] == i][columns].values)

#creacion de la grafica
fig = plt.figure()
ax = fig.add_subplot()
for count, d in enumerate(data):
    # separar las variables del grafico
    x, y = d[:,0], d[:,1]
    ax.scatter(x, y, alpha = 0.4, c = colors[count -1], edgecolors = 'none', s= 30, label = groups[count - 1])

#etiquetas
ax.set_xlabel(columns[0])
ax.set_ylabel(columns[1])
plt.legend(loc='best')
plt.title('Matplot scatter plot')
plt.show()
```

## matplotlib
- crear una lista de años de 2000 a 2020 (eje x)
- generar datos aleatorios para el eje y (hacerlo 4 veces distintas: y0, y1, y2 e y3)
- Representar las 4 secuencias de datos aleatorios en una sola figura utilizando matplotlib
- Añade una leyenda para poder identificar cada secuencia en el gráfico
- Nombra las secuencias de la siguiente forma: “Hawaii”, “San Marino”, “Islas Feroe”,
“Guayana”
- Representar los mismos datos en 4 subfiguras (axes) distintas, parte de la misma figura
- Añade la siguiente info:<br>
    – Título: “Datos aleatorios”<br>
    – Leyenda<br>
    – Nombre del eje x: “Año”<br>
    – Nombre del eje y: “GDP”<br>
- Añade al menos una anotación (texto y flecha) a uno de los gráficos


```python
import matplotlib.pyplot as plt
import numpy as np

# Se crea una lista con los años desde el 2000 hasta el 2020 para usar como base en el eje horizontal.
años = np.arange(2000, 2021)
# Se generan cuatro conjuntos diferentes de datos aleatorios para usar en el eje vertical.
y0, y1, y2, y3 = np.random.rand(4, len(años))

# Se crea una figura general donde se dibujan las cuatro líneas, asignando a cada una su etiqueta correspondiente.
plt.plot(años, y0, label="Hawaii"); plt.plot(años, y1, label="San Marino")
plt.plot(años, y2, label="Islas Feroe"); plt.plot(años, y3, label="Guayana")
# Se añade la leyenda para identificar qué color corresponde a cada lugar.
plt.legend(); plt.show()

# Se divide la figura en una rejilla de 2 por 2 para colocar cada gráfico en un espacio independiente.
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
# Se asigna un gráfico a cada recuadro con sus respectivos títulos y etiquetas para los ejes.
for ax, y, nombre in zip(axs.flat, [y0, y1, y2, y3], ["Hawaii", "San Marino", "Islas Feroe", "Guayana"]):
    ax.plot(y); ax.set_title(nombre); ax.set(xlabel="Año", ylabel="GDP")
# Se añade un título general a toda la composición y una flecha que señala un punto específico en el primer gráfico.
fig.suptitle("Datos aleatorios"); axs[0,0].annotate('Punto alto', xy=(10, 0.8), arrowprops=dict(facecolor='black'))
plt.show()
```

```python
import seaborn as sns
# Se cargan los datos directamente desde la librería de ejemplos de seaborn.
df = sns.load_dataset('mpg')

# Se crea un gráfico que muestra cómo se relacionan todas las variables numéricas entre sí.
sns.pairplot(df.dropna())

# Se muestra cómo dos variables numéricas, como peso y aceleración, tienen una relación clara mediante una línea de tendencia.
sns.lmplot(x='weight', y='acceleration', data=df)

# Se dibuja la relación entre dos variables numéricas diferenciando los colores según el origen del coche.
sns.scatterplot(x='horsepower', y='mpg', hue='origin', data=df)

# Se grafica cómo se distribuyen los valores de peso para ver promedios y puntos extremos (valores atípicos).
sns.boxplot(x=df['weight'])

# Se comparan visualmente las densidades de dos variables numéricas para ver cómo se agrupan sus valores.
sns.kdeplot(df['mpg'], fill=True); sns.kdeplot(df['displacement'], fill=True)

# Se muestra una línea de tendencia curva (orden 2) para ajustar mejor la relación entre potencia y consumo.
sns.lmplot(x='horsepower', y='mpg', data=df, order=2)
```
