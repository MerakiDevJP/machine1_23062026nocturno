#### 35. How to compute ((A+B)*(-A/2)) in place (without copy)? (★★☆)

````python
# ((A+B)*(-A/2))
# Se definen dos arrays A y B
A = np.ones(3) * 1
B = np.ones(3) * 2
print(A)
print(B)


# 1. A = A + B
A_mas_B = np.add(A, B, out=A)
print(A_mas_B)

# 2. A = A * (-A / 2)
# Nota: primero se calcula -A/2 y luego multiplicamos
A_div_dos = np.divide(A, 2, out=A)
neg_A = np.negative(A, out=A)
print(A_div_dos)
print(neg_A)

solucion = np.multiply(A, A, out=A)

# Para computar in-place, se utiliza los operadores de asignación 
# aumentada (np.add, np.divide, np.negative, etc.)
# Para evitar toda copia:
print(np.add(A, B, out=A))
print(np.divide(A, 2, out=A))
print(np.negative(A, out=A))
print(np.multiply(A, A, out=A))
````
