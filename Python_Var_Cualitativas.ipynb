```python
# Creación aleatoria de datos
import numpy as np
import random
random.seed(42) #Barajea el conjunto de datos en 42 veces
```

```python
nivel_estudios = [] # lista vacía 
nivel_estudios.extend(['Secundaria']*110)
nivel_estudios.extend(['Bachillerato']*172)
nivel_estudios.extend(['Grado']*145)
nivel_estudios.extend(['Master']*52)
nivel_estudios.extend(['Doctorado']*21)
nivel_estudios = np.random.permutation(nivel_estudios) #se genera una permutación de la lista
```

```python
print(nivel_estudios)
```

```python
# DIAGRAMA DE BARRAS
import matplotlib.pyplot as plt
import numpy as np
estudios, frecuencia = np.unique(nivel_estudios, return_counts=True) #se crean dos variables, con la función unique toma datos únicos
print(estudios, frecuencia) #se imprimen 2 listas
```

```python
plt.ylabel('Frecuencia absoluta') # etiqueta y
plt.xlabel('Nivel de estudios') # etiqueta x
plt.title('Diagrama de barras') #título del diagrama
plt.bar(estudios, frecuencia) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
plt.show()
# se debe observar si los datos son balanceados, si no es así se aplica data aumentation
```

```python
# DIAGRAMA DE SECTORES
colores = {'red', 'blue', 'orange', 'green', 'cyan'}
plt.pie(frecuencia, labels=estudios, colors=colores, autopct='%1.2f%%')
plt.title('Diagrama de sectores')
plt.show()
```

# Tarea implementar con pandas esta información

```python
import pandas as pd
import matplotlib.pyplot as plt

# se definen los datos originales
data = {
    'nivel_estudios': (['Secundaria'] * 110) + 
                      (['Bachillerato'] * 172) + 
                      (['Grado'] * 145) + 
                      (['Master'] * 52) + 
                      (['Doctorado'] * 21)
}

# Se crea el DataFrame
df = pd.DataFrame(data)

# Se verifica la creación
print("Primeras filas del DataFrame:")
print(df.head())
print("===============================================")

print("Últimas filas del DataFrame:")
print(df.tail())
print("===============================================")

print("Aleatorio de filas del DataFrame:")
print(df.sample(5))
print("===============================================")

print("Resumen concisa del DataFrame:")
df.info()
print("===============================================")

print("Tupla del DataFrame:")
print(df.shape) # es atributo no función
print("===============================================")

print("Estadística Despcriptiva del DataFrame:")
print(df.describe())
print("===============================================")

print("Estadística Despcriptiva del DataFrame:")
print(df.describe(include='all'))
print("===============================================")

print("Conteo de valores (Frecuencia de categorías):")
print(df['nivel_estudios'].value_counts())
print("===============================================")

print("Inspección directa (df):")
print(df) 
print("===============================================")

print("Nombres de las columnas:")
print(df.columns)
print("===============================================")
```

```python
# Calculamos frecuencias absolutas
frecuencias = df['nivel_estudios'].value_counts().sort_index()

# Mostramos el resumen
print("\nResumen de frecuencias:")
print(frecuencias)
```

```python
frecuencias.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Diagrama de barras (Implementado con Pandas)')
plt.ylabel('Frecuencia absoluta')
plt.xlabel('Nivel de estudios')
plt.xticks(rotation=45)
plt.show()
```

```python
frecuencias.plot(kind='pie', autopct='%1.2f%%', startangle=90, 
                 colors=['orange', 'green', 'blue', 'cyan', 'red'])
plt.title('Diagrama de sectores (Implementado con Pandas)')
plt.ylabel('') # Ocultamos el label del eje y para que se vea más limpio
plt.show()
```
